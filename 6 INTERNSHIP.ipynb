```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flipkart=pd.read_csv('flipkart_com-ecommerce_sample.csv')
```

```python
flipkart
```

```python
flipkart[flipkart.duplicated()]
```

```python
flipkart.drop_duplicates()
```

```python
flipkart.drop(['product_url','image','product_specifications'],axis=1,inplace=True)
```

```python
data1=['retail_price','discounted_price','brand','description']

for x in data1:
    flipkart[x]=flipkart[x].fillna(method='ffill')
```

```python
flipkart.isnull().sum()
```

```python
flipkart.info()
```

```python
type=['retail_price','discounted_price']
for x in type:
    flipkart[x]=flipkart[x].astype('int64')
```

```python
flipkart.nlargest(5,'retail_price')
```

```python
flipkart.nsmallest(5,'retail_price')
```

```python
plt.figure(figsize=(22,5))
flipkart.boxplot()
```

```python
f1=flipkart[flipkart['retail_price']<2500]
f1.boxplot('retail_price')
```

```python
plt.figure(figsize=(22,7))
sns.heatmap(flipkart.corr(),annot=True,cmap='rainbow')
```

```python
plt.rcParams['figure.figsize']=[20,6]
flipkart.nlargest(25,"retail_price").plot(kind="bar",x="product_name",y="retail_price",color="c",grid="True")
```

```python
plt.rcParams['figure.figsize']=[20,6]
flipkart.nsmallest(25,"retail_price").plot(kind="bar",x="product_name",y="retail_price",color="g",grid="True")
```

```python
plt.rcParams['figure.figsize']=[20,6]
flipkart.nsmallest(25,"discounted_price").plot(kind="bar",x="product_name",y="discounted_price",color="c",grid="True")
```

```python
plt.rcParams['figure.figsize']=[20,6]
flipkart.nlargest(25,"retail_price").plot(kind="bar",x="product_name",y="retail_price",color="c",grid="True")
```

```python
plt.rcParams['figure.figsize']=[20,6]
flipkart.nsmallest(25,"discounted_price").plot(kind="bar",x="brand",y="discounted_price",color="c",grid="True")
```
